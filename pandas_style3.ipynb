#### Write a Pandas program to display the dataframe in Heatmap style.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# creation of dataframe with random values
np.random.seed(42) 
data = np.random.rand(10, 4)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D'])
```

```python
#display the dataframe in Heatmap style
plt.figure(figsize=(6, 4))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Heatmap of DataFrame")
plt.show()
```
